```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
temp = data.copy()
temp['density'] = df.density
formula = 'density~C(quality)*C(alcohol)'
mod = ols(formula, data = temp).fit()
print(anova_lm(mod))
```

```python
sns.lineplot(data = temp, x = 'quality', y = 'density',hue='alcohol',ci=None);
```

```python
def combo(x):
    m = x.quality
    d = x.alcohol
    return str(m)+'/'+str(d)
```

```python
comb = temp.apply(combo,axis=1)
```

```python
print(comb)
```

```python
print(pairwise_tukeyhsd(temp.density, comb))
```
